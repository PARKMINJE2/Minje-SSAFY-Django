# ORM(Object-Relational-Mapping)

- 객체 지향 프로그래밍 언어를 사용하여 호환되지 않는 유형의 시스템 간에 데이터를 변환하는 기술

## # QuerySet API

- ORM에서 데이터를 검색, 필터링, 정렬 및 그룹화 하는 데 사용하는 도구

	- API를 사용하여 SQL이 아닌 Python 코드로 데이터를 처리

#### # QuerySet



- QuerySet API는 python의 모델 클래스와 인스턴스를 활용해 DB에 데이터를 저장, 조회, 수정, 삭제(CRUD)하는 것

### # CRUD

- 소프트웨어가 가지는 기본적인 데이터 처리 기능

	- 'C'reate (저장)

	- 'R'ead (조회)

	- 'U'pdate (갱신)

	- 'D'elete (삭제)

#### 1. 데이터 객체를 만드는(생성하는) 3가지 방법 - "C"reate

- #첫번째 방법





	- pk: primary key(권장)



- #두번째 방법

	- save 메서드를 호출해야 비로소 DB에 데이터가 저장됨

	- 테이블에 한 행(레코드)이 쓰여진 것



- #세번째 방법

	- QuerySet API 중 create() 메서드 활용



---

#### 2. "R"ead

- 대표적인 조회 메서드



	- all()

: 전체 데이터를 조회



	- filter()

: 주어진 매개변수와 일치하는 객체를 포함하는 "QuerySet" 반환



	- get()

: 주어진 매개변수와 일치하는 객체를 반환(없는거, 두개인거 조회 못함)



객체를 찾을 수 없으면 DoesNotExist 예외를 발생시키고, 둘 이상의 객체를 찾으면 MultipleObjectsReturned 예외를 발생시킴. 그러므로 "primary key와 같이 고유성을 보장하는 조회에서 사용"해야 함.

---

#### # 3. "U"pdate

- 데이터 수정

	- 조회하기 -> 수정하기 -> 저장하기

	- 인스턴스 변수를 변경 후 save 메서드 호출



---

#### # 4. "D"elete

- 데이터 삭제

	- 삭제하려는 데이터 조회 후 delete 메서드 호출
